#  <font color='blue'> 1. Importing Required Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt 
import sklearn as sk
```

# <font color='blue'> 2. Reading the CSV file

```python
df=pd.read_csv(r"C:\Users\userkk\Downloads\Kodi\Student mark prediction/Student_Marks.csv")
```

# <font color='blue'> 3. Checking the null count

```python
df.isnull().sum()
```

```python
df.columns
```

```python
X=df[['number_courses', 'time_study']]
y=df["Marks"]
```

# <font color='blue'> 4. Splitting the dataset into train test split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=22)
```

# <font color='blue'> 5. Evaluating Our Model

# <font color='purple'> A. Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
y_predict=model.predict(X_test)
```

## <font color='green'> Accuracy of the model :-

```python
from sklearn.metrics import r2_score
print(r2_score(y_test,y_predict))
```

# <font color='purple'> B. RandomForest Regression :-

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor()
model.fit(X_train,y_train)
```

```python
y_pre=model.predict(X_test)
```

## <font color='green'> Accuracy of the model :-

```python
from sklearn.metrics import r2_score
print(r2_score(y_test,y_pre))
```

# <font color='purple'> C. KNeighrest Neighbour :-

```python
from sklearn.neighbors import KNeighborsRegressor
model=KNeighborsRegressor()
model.fit(X_train,y_train)
```

```python
y_pr=model.predict(X_test)
```

## <font color='green'> Accuracy of the model :-

```python
from sklearn.metrics import r2_score
print(r2_score(y_test,y_pr))
```

```python
print("Accuracy of Linear Regression is :- ",r2_score(y_test,y_predict))
print("Accuracy of RandomForestRegressor is :- ",r2_score(y_test,y_pre))
print("Accuracy of KNN is :- ",r2_score(y_test,y_pr))
```

<b> It is found that the KNN algorithm has the more accuracy as compared to the linearregression and randomforestregressor. </b>
